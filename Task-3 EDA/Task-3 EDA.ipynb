```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "C:\\Users\\91739\\Downloads\\Telco_Customer_Churn_Dataset  (3).csv"  
df = pd.read_csv(file_path)
```

```python
# EXPLORATORY DATA ANALYSIS (EDA)
print("\nStarting Exploratory Data Analysis...")
```

```python
print("\nSummary Statistics:")
display(df.describe())
```

```python
# Churn distribution
if 'churn' in df.columns:
    print("\nChurn Distribution:")
    plt.figure(figsize=(6,4))
    sns.countplot(x='churn', data=df, palette='coolwarm')
    plt.title('Churn Distribution')
    plt.show()

```

```python
# Tenure distribution
print("\nTenure Distribution:")
sns.histplot(df['tenure'], bins=30, kde=True, color='blue')
plt.title('Tenure Distribution')
plt.xlabel('Tenure (Months)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Monthly charges distribution
print("\nMonthly_Charges Distribution:")
sns.histplot(df['MonthlyCharges'], bins=30, kde=True, color='green')
plt.title('Monthly_Charges Distribution')
plt.xlabel('Monthly_Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of Monthly Charges by Churn
if 'churn' in df.columns and 'monthlycharges' in df.columns:
    print("\nBox Plot: Monthly_Charges by Churn")
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df['churn'].astype(str), y=df['Monthly_Charges'], data=df, palette='coolwarm')
    plt.title('Monthly_Charges by Churn')
    plt.xlabel('Churn')
    plt.ylabel('Monthly_Charges')
    plt.show()
```
